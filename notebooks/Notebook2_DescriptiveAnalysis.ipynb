##  Summary of Findings: Descriptive Analysis of Proxy Maternal Health Funding

This notebook explores the distribution and trends in **proxy maternal health funding** across Texas counties from 2016 to 2023. The proxy funding index was calculated using the formula:

\[
\text{Proxy Funding} = \text{Number of Births} \times \$39.34
\]

###  Key Insights:

- **Top Counties by Average Funding**  
  Rockwall, Coryell, and Maverick counties received the highest average funding, despite not being the most populous. This suggests that mid-sized counties with consistent birth volumes may be strong beneficiaries of per-birth-based funding models.

- **Funding Trends Over Time**  
  Annual funding fluctuated between 2017 and 2021, with notable declines in 2018 and 2020, likely reflecting drops in birth rates during those years. This highlights how maternal funding  if tied to birth counts  may be sensitive to demographic and external factors (e.g., the COVID-19 pandemic).

- **Strong Correlation Between Births and Funding**  
  A near-perfect linear relationship was observed between number of births and proxy funding. This validates the index design and confirms that the funding calculation was applied accurately across all counties and years.

###  Implications:
- Counties with consistent birth volumes benefit more predictably from funding mechanisms based on volume.
- Rural or low-birth counties may be underfunded relative to need  a consideration for equity-based funding policy adjustments.
- These descriptive outputs provide a foundation for deeper analysis in the next notebook, where proxy funding will be compared with maternal health outcomes like preterm birth and infant mortality.

```python
import pandas as pd
import glob
import os

# Load all CSVs from correct folder
csv_files = glob.glob("/Users/stephmozley/Desktop/UTH SBMI/BMI 6000 Practicum/DSHS/raw_data/texas_live_births_*.csv")
print(f"Found {len(csv_files)} files")

df_list = []

for file in csv_files:
    year = os.path.basename(file).split("_")[-1].split(".")[0]
    try:
        df = pd.read_csv(file, sep="\t", encoding="utf-8", on_bad_lines="skip")
    except UnicodeDecodeError:
        df = pd.read_csv(file, sep="\t", encoding="utf-16", on_bad_lines="skip")
    
    df["Year"] = int(year)
    df_list.append(df)

# Merge
births_df = pd.concat(df_list, ignore_index=True)

# Rename and clean
births_df.rename(columns={"Category": "County", "suppressed": "LiveBirths"}, inplace=True)
births_df["County"] = births_df["County"].str.strip()
births_df["LiveBirths"] = pd.to_numeric(births_df["LiveBirths"], errors="coerce")

# Preview
births_df.head()
```

```python
# Aggregate live births by county and year
births_summary = births_df.groupby(["County", "Year"], as_index=False)["LiveBirths"].sum()

# Average annual births per county (20172021)
births_avg = births_summary.groupby("County", as_index=False)["LiveBirths"].mean()
births_avg.rename(columns={"LiveBirths": "AvgAnnualBirths"}, inplace=True)

# Merge with geodata
gdf = gdf.merge(births_avg, on="County", how="left")
```

```python
gdf["ProxyFunding"] = gdf["AvgAnnualBirths"] * 100  # change multiplier as needed
```

 Statistical Interpretation of the Correlation
 Strength and Direction
r = 0.83 indicates a strong positive linear correlation.

This means that as the number of uninsured individuals in a county increases, the estimated proxy maternal health funding tends to increase in a consistent and predictable way.

 Statistical Significance
p-value = 0.0000 (technically, p < 0.0001) means the result is highly statistically significant.

You can confidently reject the null hypothesis of no correlation  this relationship is not due to random chance.

 Practical Implications
The strong correlation validates your funding proxy: birth counts scaled by per-birth allocations align well with actual indicators of maternal health need (like lack of insurance).

Counties with high uninsured rates generally receive higher funding estimates under your model.

This suggests your model could be a useful tool for targeting maternal health investments, especially where detailed funding data is not available.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Drop NaNs for modeling
plot_df = plot_df.dropna(subset=["Uninsured", "ProxyFunding"]).copy()

# Reshape for sklearn
X = plot_df["Uninsured"].values.reshape(-1, 1)
y = plot_df["ProxyFunding"].values

# Fit linear model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Residuals
plot_df["Residual"] = y - y_pred

# Flag outliers: top 5% lowest residuals (underfunded)
threshold = np.percentile(plot_df["Residual"], 5)
plot_df["Outlier"] = plot_df["Residual"] < threshold

# Plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=plot_df, x="Uninsured", y="ProxyFunding", hue="Outlier", palette={False: "blue", True: "red"}, alpha=0.7)

# Label top outliers
for _, row in plot_df[plot_df["Outlier"]].iterrows():
    plt.text(row["Uninsured"], row["ProxyFunding"], row["County"].title(), fontsize=8)

plt.title("Outlier Counties: Uninsured vs Proxy Maternal Health Funding (2023)")
plt.xlabel("Number of Uninsured Individuals")
plt.ylabel("Estimated Proxy Maternal Health Funding")
plt.legend(title="Underfunded Outlier", loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```
